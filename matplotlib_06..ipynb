```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%", radius=1.5)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%", radius=1.5, shadow=True)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%", radius=1.5, shadow=True, explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%", radius=1.5, shadow=True, explode=[0,0.2,0,0,0],startangle=90)
plt.show()
```
