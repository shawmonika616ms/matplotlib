```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car","Other utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,autopct="%0.2f%%", radius=1.5, shadow=True, explode=[0,0.2,0,0,0],startangle=90)
plt.show()
```

```python
plt.savefig("c:/datas/piechart.png", bbox_inches="tight",pad_inches=2,transparent=True)
```
